```python
import pandas as pd
import re
import contractions
```

```python
df = pd.read_csv('labeled_data.csv', on_bad_lines='skip')
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
df = df[df['class'] == 0]
```

```python
tweet_text = df['tweet'] # convert the tweets into lower case
tweet_text = tweet_text.tolist() # create a list of tweets

# remove emojis from the tweets
def remove_emoji(s):
    emoji_pattern = re.compile(
          '['
          u'\U0001F600-\U0001F64F'  # emoticons
          u'\U0001F300-\U0001F5FF'  # symbols & pictographs
          u'\U0001F680-\U0001F6FF'  # transport & map symbols
          u'\U0001F1E0-\U0001F1FF'  # flags (iOS)
          u'\U00002702-\U000027B0'
          u'\U000024C2-\U0001F251'
          ']+',
          flags=re.UNICODE)
    return emoji_pattern.sub(r'', s)

# remove URL from the tweets
def remove_URL(s):
    url = re.compile(r'https?://\S+|www\.||pic.twitter.com\S+')
    return url.sub(r'', s)

# to remove any special characters and extra space
def normalizeString(s): 
    s = re.sub("@[A-Za-z0-9]+","",s)
    s = contractions.fix(s)
    s = remove_emoji(s)
    s = remove_URL(s)
    s = re.sub(r"[^a-zA-Z0-9.,?' ]+", r" ", s)
    s = s.strip()
    
    return s

# cleaning tweets in the dataset
for i in range(len(tweet_text)):
    tweet_text[i] = str(tweet_text[i])
    tweet_text[i] = ' '.join(tweet_text[i].split('\n'))
    tweet_text[i] = tweet_text[i].replace('&amp', '')
    words = filter(lambda x: x != "rt", tweet_text[i].split())
    tweet_text[i] = ' '.join(words)
    tweet_text[i] = ' '.join(normalizeString(str).strip() for str in tweet_text[i].split())
```

```python
df['tweet_text'] = tweet_text
```

```python
all_text = '\\'.join([text for text in df['tweet_text']])
```

```python
tagged_sentence = nltk.tag.pos_tag(all_text.split())
```

```python
edited_sentence = [word for word,tag in tagged_sentence if tag != 'NNP' and tag != 'NNPS']
```

```python
all_text = ' '.join(edited_sentence)
all_text = all_text.lower()
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) # stop words in english
stop_words = stop_words.union(['like', 'one', 'people', 'low', 'call', 'smh', 'realdonaldtrump', 'get', 'know', 'got', 'get', 'man', 'would', 'say', 'want'])
```

```python
import yake

kw_extractor = yake.KeywordExtractor()

max_ngram_size = 3

deduplication_threshold = 0.5

numOfKeywords = 50

custom_kw_extractor = yake.KeywordExtractor(lan='en', top=numOfKeywords, features=None, stopwords = stop_words)

keywords = custom_kw_extractor.extract_keywords(all_text)

for kw in keywords:
     print(kw)
```
